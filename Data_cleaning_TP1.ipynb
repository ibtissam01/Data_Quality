2.2 Traitement des valeurs manquantes

2.2.1 Remplacer les valeurs manquantes de manière simple

Method to use for filling holes in reindexed Series:


ffill: propagate last valid observation forward to next valid.


backfill / bfill: use next valid observation to fill gap.

```python
df1 = data.copy() #dataframe copy
```

```python
df1["cabin"].fillna(method='pad', inplace = True)
```

```python
df1["body"].fillna(method='bfill', inplace = True)
```

```python
data.isna().sum()
```

```python
df1.isna().sum()
```

2.2.2 Traitement des valeurs manquantes par imputation statistique :

```python
df2 = data.copy() #dataframe copy
```

```python
from sklearn.impute import SimpleImputer
```

```python
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
df2['age'] = imp.fit_transform(df2['age'].values.reshape(-1,1))
df2.isnull().sum()
```
